```python
import collections
import numpy as np
import seaborn as sns
import os
import matplotlib.gridspec as gridspec
import pickle
import matplotlib as mpl
from pathlib import Path

pgf_with_custom_preamble = {
    "text.usetex": False,    # use inline math for ticks
    "pgf.rcfonts": False,   # don't setup fonts from rc parameters
    "pgf.preamble": [
        '\\usepackage[T1]{fontenc}'
     ]
}


sns.set_style('ticks')
sns.set_context('poster')
sns.set_palette('Set2', 25)

colors = sns.color_palette('Set2', 20)

from scipy import interpolate
```

```python
polstats = {}
path = Path('with_water_removed/')
for f in path.iterdir():
    if 'polstat' in f.name:
        k = float(f.name.split('_')[2])
        polstats[k] = np.loadtxt(f)
        with f.open('r') as h:
            header = h.readline().replace('# ', '').split()
            polstats[k].dtype = [(x, np.float) for x in header]
```

```python
for k in polstats:
    plt.plot(polstats[k]['cr'], polstats[k]['pdi'], '*')
```
